```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver_path = 'chromedriver-win64/chromedriver-win64/chromedriver.exe'  # 替换为您 ChromeDriver 的实际路径
driver = webdriver.Chrome(executable_path=driver_path)

# 打开目标网页
driver.get('https://drugs.dxy.cn/pc/clinicalDecision/KyuVVvxdHam6FdSMumepepmLzWQ==')  # 替换为实际的网址

# 等待页面加载完成
time.sleep(2)  # 可以根据需要调整等待时间

# 定位所有 data-menu-index 属性为 '2' 的 div 元素
target_divs = driver.find_elements(By.XPATH, "//div[@data-menu-index='2']")

# 遍历所有找到的 div 元素
for target_div in target_divs:
    try:
        # 获取该 div 下的第一个 h2 标签的文字内容
        h2_text = target_div.find_element(By.TAG_NAME, 'h2').text
        print(f"H2 Text: {h2_text}")
    except:
        print("No h2 tag found in this div")

    # 获取该 div 下的所有 div 标签的文字内容
    child_divs = target_div.find_elements(By.TAG_NAME, 'div')
    for child_div in child_divs:
        child_text = child_div.text
        print(f"Child Div Text: {child_text}")

# 关闭浏览器
# driver.quit()
```
